```python
base_url = "https://makercentral.io/levels/search/_?difficulty=Normal&clearStatus=Cleared&minimumPlays=1000&page="
level_data_url = "https://tgrcode.com/mm2/level_data/"
levels_directory = "D:\\smm2\\levels"

response = requests.get(base_url + "0")
soup = BeautifulSoup(response.content, "html.parser")

# Lista para guardar los archivos descargados
downloaded_files = []

# Busca el span que contiene "Found about"
span_element = soup.find('span', string=re.compile("Found about"))

# Extrae y imprime la cadena después de "Found about"
if span_element:
    result_text = span_element.string
    found_index = result_text.find("Found about")
    if found_index != -1:
        result_text = result_text[found_index + len("Found about"):].strip()
        results_index = result_text.find(" results")
        if results_index != -1:
            numero_resultados = int(result_text[:results_index].replace(",", ""))
            numero_paginas = numero_resultados // 10
            residuo = numero_resultados % 10
            numero_resultados_ult_pagina = residuo if residuo != 0 else 10
            print("Numero de resultados:", numero_resultados)
            print("Numero de paginas:", numero_paginas)
            print("Numero de resultados en la última página:", numero_resultados_ult_pagina)
            
            for i in range(20):
                random_num = random.randint(1, numero_resultados)
                pagina = (random_num - 1) // 10
                posicion = (random_num - 1) % 10 + 1

                print(f"Número aleatorio: {random_num}, en la página: {pagina}, posición: {posicion}")
                
                level_code = None
                retries = 0

                while level_code is None and retries < 2:
                    response = requests.get(base_url + str(pagina))
                    soup = BeautifulSoup(response.content, "html.parser")
                    level_previews = soup.find_all('div', class_='user-level-preview-wrapper')

                    if posicion - 1 < len(level_previews):
                        level_link = level_previews[posicion - 1].find('a', href=re.compile(r'/levels/view/'))
                        if level_link:
                            level_code = re.search(r'/levels/view/(.+)', level_link['href']).group(1)
                    
                    if level_code is None:
                        time.sleep(5)
                        retries += 1
                
                if level_code is not None:
                    print(f"Nivel {i+1}: {level_code}")
                    download_url = level_data_url + level_code
                    download_response = requests.get(download_url)
                    file_path = os.path.join(levels_directory, f"course_data_{str(i).zfill(3)}.bcd")
                    with open(file_path, 'wb') as file:
                        file.write(download_response.content)
                    print(f"Archivo descargado en: {file_path}")
                    downloaded_files.append(file_path)
                else:
                    print(f"No se pudo obtener el código del nivel para el número aleatorio {random_num}")

            # Verificación y reintentos de descarga, si es necesario
            for file_path in downloaded_files:
                download_retries = 0
                expected_file_size = 368 * 1024

                while download_retries < 3:
                    actual_file_size = os.path.getsize(file_path)
                    if actual_file_size == expected_file_size:
                        break

                    print(f"El tamaño del archivo {file_path} no coincide. Esperado: {expected_file_size}, actual: {actual_file_size}. Intentando nuevamente...")
                    
                    level_code = os.path.basename(file_path).split("_")[2].split(".")[0]
                    download_url = level_data_url + level_code
                    download_response = requests.get(download_url)
                    with open(file_path, 'wb') as file:
                        file.write(download_response.content)
                    
                    download_retries += 1
                    time.sleep(0 if download_retries == 1 else (5 if download_retries == 2 else 10))
                
                if download_retries == 3:
                    print(f"No se pudo descargar el archivo {file_path} correctamente después de {download_retries} intentos")
```
